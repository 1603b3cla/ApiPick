Para formalizar el modelo de McCulloch y Pitts, representaremos una neurona con la letra  C  y definiremos que la entrada de dicha neurona, denominada  input , está determinada por una tupla  (E,I) , donde  E  es el conjunto de señales excitatorias e  I  es el conjunto de señales inhibitorias. Asimismo, estableceremos como restricciones que tanto la entrada como la salida sean variables binarias, es decir,  input,output∈0,1 . Adicionalmente, definiremos un umbral  u  que la neurona utiliza para determinar los casos de excitación..

Dado lo anterior la salida de  C  se calcula usando las siguientes reglas:

En caso de que alguna de las entradas inhibitorias esté activa la neurona no se exita, es decir,

C(E,I)=0  & \texttt{if} &  ∃ιi=1;ιi∈I \


La neurona se exita si la integral de sus entradas exitatorias es igual o superior al umbral, es decir,

C(E,I)=1  & \texttt{if} &  ∑ιi≥u;ιi∈I \

En cualquier otro caso la neurona permanece sin exitación.

```python
import numpy as np
```

```python
def neuronaMyP(E,I,u):
    for ini   in I:
      if ini == 1:
        return 0

        integracion=0
        for ex in E:
          integracion = integracion + ex

        if integracion >= u:
          return 1
        else:
          return 0
```

```python
E = [1]
I = [0]
u = 1

print(neuronaMyP(E,I,u))
```

Conclusión

La neurona artificial implementada en el código anterior es una neurona de McCulloch-Pitts, que es un modelo simple de neurona que se utiliza para comprender los principios básicos del procesamiento de la información en el cerebro.

La neurona recibe dos tipos de entradas: excitatorias (E) e inhibitorias (I). Las entradas excitatorias aumentan la probabilidad de que la neurona se dispare, mientras que las entradas inhibitorias disminuyen la probabilidad de que la neurona se dispare.

El umbral (u) es el valor mínimo de integración que debe alcanzarse para que la neurona se dispare.

En el ejemplo proporcionado, la neurona tiene una entrada excitatoria de 1 y una entrada inhibitoria de 0. El umbral es 1.

Dado que la integración de la entrada excitatoria (1) es mayor que el umbral (1), la neurona se disparará.

Este ejemplo muestra cómo una neurona artificial puede procesar entradas excitatorias e inhibitorias para tomar una decisión sobre si disparar o no.
